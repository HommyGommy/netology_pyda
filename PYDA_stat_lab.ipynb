# Вагоны и седаны стоят одинаково

```python
p_wagon = df[ df['body-style']=='wagon']['price']
p_sedan = df[ df['body-style']=='sedan']['price']

len(p_wagon), len(p_sedan)
```

```python
p_wagon = p_wagon.sample(25)
p_sedan = p_sedan.sample(25)
```

```python
statistics.variance(p_wagon), statistics.variance(p_sedan)
```

```python
sts.ttest_ind(p_wagon, p_sedan, equal_var=False)
```
